## Step 4: Run a simple regression in Statsmodels with TV as a predictor

```python
# import libraries
import statsmodels.api as sm
from statsmodels.formula.api import ols

# build the formula 
formula = 'sales~TV'

# create a fitted model in one line
model = ols(formula, data=df).fit()
```

## Step 5: Get Regression Diagnostics Summary

```python
model.summary()
```

Note here that the coefficients represent associations, not causations

## Step 6:  Draw a prediction line with data points on a scatter plot for X (TV) and Y (Sales)

Hint: You can use the `model.predict()` function to predict the start and end point of of regression line for the minimum and maximum values in the 'TV' variable. 

```python
# create a DataFrame with the minimum and maximum values of TV
X_new = pd.DataFrame({'TV': [df.TV.min(), df.TV.max()]})
print(X_new.head())

# make predictions for those x values and store them
predicted_values = model.predict(X_new)
print(predicted_values)
# first, plot the observed data and the least squares line
plt.scatter(df.TV, df.sales)
plt.plot(X_new, predicted_values, color='red')
plt.show()
```

## Step 7: Visualize the error term for variance and heteroscedasticity

```python
fig = plt.figure(figsize=(40,20))
fig = sm.graphics.plot_regress_exog(model, 'TV', fig=fig)
plt.show()
```

```python
# Record Your observations on heteroscedasticity
#The errors are heteroskedastic. The residuals vs TV plot shows the variance in the residuals increases as the value
#of TV increases. 

#The residuals plot above demonstrates heteroscedasticity. The residuals plot displays a cone shape indicating that 
#as values of the independent variable increases in value the variance of the residuals increases.
```

## Step 8: Check the normality assumptions by creating a QQ-plot

```python
# Code for QQ-plot here
resid1 = model.resid
fig = sm.graphics.qqplot(resid1, dist=stats.norm, line='45', fit=True)
fig.show()
```

```python
# Record Your observations on the normality assumption
#The normality assumption holds as evidence by the conformance of blue dots to the red line in the Q-Q plot above. 
```

## Step 9: Repeat the above for radio and record your observations

```python
# code for model, prediction line plot, heteroscedasticity check and QQ normality check here
# build the formula 
formula2 = 'sales~radio'

# create a fitted model in one line
model2 = ols(formula2, data=df).fit()
```

```python
model2.summary()
```

```python
# create a DataFrame with the minimum and maximum values of TV
X_new2 = pd.DataFrame({'radio': [df.radio.min(), df.radio.max()]})
print(X_new2.head())

# make predictions for those x values and store them
predicted_values2 = model2.predict(X_new2)
print(predicted_values2)
# first, plot the observed data and the least squares line
plt.scatter(df.radio, df.sales)
plt.plot(X_new2, predicted_values2, color='red')
plt.show()

```

```python
fig2 = plt.figure(figsize=(40,20))
fig2 = sm.graphics.plot_regress_exog(model2, 'radio', fig=fig2)
plt.show()
```

```python
resid2 = model2.resid
fig = sm.graphics.qqplot(resid2, dist=stats.norm, line='45', fit=True)
fig.show()
```

```python
# Record your observations here for goodnes of fit 
#The R^2 is .33 and the adjusted R^2 is .32 indicating that the model accounts for approximately one third of 
#the variance in the dependent variable. The residuals, however, are heteroskedastic as demonstrated in the Residuals VS Radio plot.
#The residuals are not normally distributed, as indcated by the Q-Q plot and the Jarque-Bera test result (JB = 21.91, p = 0.000).
```
